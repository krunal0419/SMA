```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Tweets.csv")

# Display the first few rows of the dataset
df.head()

# Visualize tweet counts over time
df['tweet_created'] = pd.to_datetime(df['tweet_created'])  # Convert Date column to datetime format
plt.figure(figsize=(10, 6))
df.resample('D', on='tweet_created').size().plot(kind='line')
plt.title('Number of Tweets about Artificial Intelligence over Time')
plt.xlabel('tweet_created')
plt.ylabel('Number of Tweets')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
df[['airline_sentiment', 'airline_sentiment_confidence']].plot(kind='box')
plt.title('Sentiment of customers')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```
