```python
import pandas as pd
from textblob import TextBlob

# Load dataset
df = pd.read_csv("Twitter.csv")

# Extract tweets from the dataset column
tweets_column = "tweet"  # Replace with the name of the column containing tweets
tweets = df[tweets_column].tolist()

# Analyze sentiment of each tweet
sentiments = []
polarities = []
for tweet in tweets:
    analysis = TextBlob(str(tweet))
    polarity = analysis.sentiment.polarity
    polarities.append(polarity)
    if polarity > 0:
        sentiments.append("Positive")
    elif polarity < 0:
        sentiments.append("Negative")
    else:
        sentiments.append("Neutral")

# Add sentiment analysis results to the dataframe
df["Polarity"] = polarities
df["Sentiment"] = sentiments

# Display sentiment analysis results
df.head() # Display the first few rows of the dataframe with sentiment analysis results
```
